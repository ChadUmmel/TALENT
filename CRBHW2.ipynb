```python
## This code produces a plot to solve Homework 2 assigned by Carl Brune to the TALENT Nuclear Reaction Theory
#Summer School in 2019
#Authors:  Chad Ummel and Rajesh Ghimire

import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Physical constants
hbar = 197.3286 #(hbar*c in MeV fm)
m16O = 15.99491461956*931.5 #Alpha mass (in MeV)
mn = 939.565 #Neutron mass (in MeV)
mu = m16O*mn/(mn+m16O) #Reduced mass
```

```python
#Constants defined by the problem
ER = 2.212
gamma2 = 0.04
Gamma = 0.1268
a = 5
```

```python
#Compute the phaseshift
def simpleR(E):
    k = np.sqrt(2*mu*E)/hbar
    rho = k*a
    delta = np.arctan2(rho*gamma2,(ER-E))-rho
    deltaBW = np.arctan2(Gamma/2,(ER-E))
    return(k,delta,deltaBW)
```

```python
#Compute the cross section
def crosssection(E):
    (k,delta,deltaBW)=simpleR(E)
    cs = 4*math.pi/(k**2)*(np.sin(delta))**2/100
    csBW = 4*math.pi/(k**2)*(np.sin(deltaBW))**2/100
    return(cs,csBW)
```

```python
#Get the phase shift as a function of E
E = np.linspace(0,4,401)
(k,delta,deltaBW) = simpleR(E)
plt.plot(E,delta/math.pi*180,label="R-matrix")
plt.plot(E,deltaBW/math.pi*180,label="Breit-Wigner")
plt.grid()
plt.xlabel("E_CM (MeV)")
plt.ylabel("phase shift (degrees)")
plt.xlim(0,)
plt.legend();
```

```python
#Get the cross section as a function of E
E = np.linspace(.01,4,401)
(cs,csBW)=crosssection(E)
plt.plot(E,cs,label="R-matrix")
plt.plot(E,csBW,label="Breit-Wigner")
plt.grid()
plt.xlabel("E_CM (MeV)")
plt.ylabel("Cross Section")
plt.xlim(0,)
plt.ylim(0,)
plt.legend();
print("Look how incorrect Breit-Wigner is")
```
