```python
#This code produces a plot to solve Problem 3 of Homework 1 assigned by Carl Brune to the TALENT Nuclear Reaction Theory
#Summer School in 2019
#Authors:  Chad Ummel and Rajesh Ghimire

import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Physical constants
hbar = 197.3286 #(hbar*c in MeV fm)
ma = 3727.79378 #Alpha mass (in MeV)
mn = 939.565 #Neutron mass (in MeV)
mu = ma*mn/(mn+ma) #Reduced mass
```

```python
#Constants defined by the problem
ER = 50
gamma2 = 10
a = 3
```

```python
#Compute the first-step variables
def simpleR(E):
    k = np.sqrt(2*mu*E)/hbar
    rho = k*a
    delta = np.arctan(rho*gamma2/(ER-E))-rho
    return(delta)
```

```python
#Get the phase shift as a function of E
E = np.linspace(0,20,201)
delta = simpleR(E)
plt.plot(E,(delta+math.pi)/math.pi*180)
plt.grid()
plt.xlabel("E_CM (MeV)")
plt.ylabel("phase shift (degrees)")
plt.xlim(0,);
```
